```python
import tensorflow as tf
from tensorflow.python.platform import gfile
import numpy as np
import time
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
```

```python
model_file = 'R:/ONNX/Main/tensorflow_inception_graph.pb'
```

```python
with tf.compat.v1.Session() as sess:
    with gfile.FastGFile(model_file,'rb') as f:
        graph_def = tf.compat.v1.GraphDef()
        graph_def.ParseFromString(f.read())
        tf.import_graph_def(graph_def, name='')
   

    print('Tensorflow model loaded')
    model_info = [op.name for op in tf.compat.v1.get_default_graph().get_operations()]
    print('Model info:')
    #print(model_info)


    print('Model layer info:')
   # print([op.values() for op in tf.compat.v1.get_default_graph().get_operations()])

    print('Time taken by each layer to execute:')
    for op in tf.compat.v1.get_default_graph().get_operations():
        print(op.values())


    print('Time taken by the model to execute:')
    start_time = time.time()
    for i in range(100):
        image_data = tf.compat.v1.gfile.FastGFile('C:/Users/gouru/Pictures/Screenshots/Screenshot (1).png','rb').read()
        logits = tf.compat.v1.get_default_graph().get_tensor_by_name('conv2d0_w:0')
        predictions = sess.run(logits,{'mixed5b_1x1_pre_relu/conv':image_data})
    print(time.time()-start_time)
```

```python
output_index = [op.values() for op in tf.compat.v1.get_default_graph().get_operations()].index('InceptionV3/Logits/SpatialSqueeze') + 1
```

```python
output_index = np.argmax(predictions[0])
print('Prediction:')
print(output_index)
```
