```python
import onnx
import onnxruntime
import numpy as np
import sys
```

```python

model_file = sys.argv[1]

model = onnx.load('R:/ONNX/resnet50-v1-7.onnx')
print(model.graph.input)
print(model.graph.output)
```

```python
output_name = sess.get_outputs()[0].name
print(output_name)
print(sess.get_outputs()[0].shape)
```
