```python
import tensorflow as tf
import time
```

```python
with tf.device('/cpu:0'):
    model = tf.keras.applications.mobilenet_v2.MobileNetV2(input_shape=(224,224,3), include_top=False)

model.summary()
```

```python

start_time = time.time()
x = model.layers[0].output
for i in range(1, len(model.layers)):
    x = model.layers[i].output
    end_time = time.time()
```

```python

print(f"Total number of layers: {len(model.layers)}")
print(f"Time taken to execute first layer: {end_time - start_time}")
```

```python

start_time = time.time()
x = model.layers[1].output
for i in range(2, len(model.layers)):
    x = model.layers[i].output
    end_time = time.time()

print(f"Time taken to execute second layer: {end_time - start_time}")
print(f"Model name: {model.layers[i].name}")
```

```python
start_time = time.time()
x = model.layers[0].output
for i in range(1, len(model.layers)):
    x = model.layers[i].output
    end_time = time.time()
    print(f"Time taken to execute layer {i}: {end_time - start_time}")
    print(f"Model name: {model.layers[i].name}")
```

```python
start_time = time.time()
x = model.layers[0].output
for i in range(1, len(model.layers)):
    x = model.layers[i].output
    end_time = time.time()

total_time = 0
for i in range(1, len(model.layers)):
    total_time += end_time - start_time

print(f"Total time taken to execute the model: {total_time}")
```
